```python
import math
import matplotlib.pyplot as plt

# Constants
motor_spacing = 90  # in mm

# Given coordinates for the wave motion
# Replace these coordinates with the actual coordinates you have
# Format: (x, y)
coordinates = [(0, 0),(90,-21.4),
    (180, -13.23),
    (270, 13.23),
    (360, 21.4),

]

# Example equation of the wave motion (You should replace this with your own equation)
def wave_equation(x):
    return math.sin(2 * x)  # Example: a simple sine wave function

# Function to calculate the angle for servo motors
def calculate_servo_angles(coords):
    angles = []
    for x, y in coords:
        # Calculate the height of the wave at the given x coordinate using the equation
        wave_height = wave_equation(x)

        # Calculate the distance between each point and its previous one
        distance = math.sqrt(x ** 2 + y ** 2)

        # Calculate the angle for the servo based on the wave height
        if distance != 0:
            angle = math.atan(wave_height / distance)
            servo_angle = math.degrees(angle)
            angles.append(servo_angle)
        else:
            angles.append(0)  # Initial position for the first servo

    return angles

# Calculate servo angles
servo_angles = calculate_servo_angles(coordinates)

# Plotting the wave motion graph
x_values = [coord[0] for coord in coordinates]
y_values = [wave_equation(x) for x in x_values]

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='Wave Motion')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Wave Motion Graph')

# Plot angles on the graph
for i, (x, y) in enumerate(zip(x_values, y_values)):
    plt.text(x, y, f'{servo_angles[i]}', ha='right', va='bottom')

plt.legend()
plt.grid(True)
plt.show()
```
